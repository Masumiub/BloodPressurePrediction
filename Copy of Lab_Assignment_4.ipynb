```python
# write your code here
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
```

```python
w = np.array([44, 61, 81, 113, 131])
p = np.array([91, 98, 103, 110, 112])
```

```python
#x = np.log(w)
arr = []
for i in w:
  arr.append(math.log(i))

arr =  np.array(arr)
```

```python
#X = np.column_stack((np.ones_like(x), x))
w_avg = np.mean(arr)
p_avg = np.mean(p)
```

```python
#beta_hat = np.linalg.lstsq(X, p, rcond=None)[0]

w_demean = arr - w_avg
p_demean = p - p_avg

w_std = np.std(w_demean)
p_std = np.std(p_demean)
```

```python
rho = np.dot(w_demean, p_demean.T)/np.sqrt(np.sum(w_demean**2) * np.sum(p_demean**2))
```

```python
β1 = rho * p_std/w_std
β0 = p_avg - w_avg*β1
```

```python

def bloodpressure(w):
  x = math.log(w)
  return β0+β1 * x

print("Predicted systolic blood pressure for a healthy child weighing 100 pounds:", bloodpressure(100))
```
